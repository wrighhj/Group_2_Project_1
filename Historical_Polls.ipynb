```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from scipy.stats import linregress

# Historical Poll Info: source https://www.presidency.ucsb.edu/statistics/data/presidential-job-approval

trump = pd.read_csv("Historical_Polls/trump.csv")
obama = pd.read_csv("Historical_Polls/obama.csv")
bush_w = pd.read_csv("Historical_Polls/bush_w.csv")
clinton = pd.read_csv("Historical_Polls/clinton.csv")
bush_hw = pd.read_csv("Historical_Polls/bush_hw.csv")
reagan = pd.read_csv("Historical_Polls/reagan.csv")
carter = pd.read_csv("Historical_Polls/carter.csv")
ford = pd.read_csv("Historical_Polls/ford.csv")
nixon = pd.read_csv("Historical_Polls/nixon.csv")
johnson = pd.read_csv("Historical_Polls/johnson.csv")
kennedy = pd.read_csv("Historical_Polls/kennedy.csv")
eisenhower = pd.read_csv("Historical_Polls/eisenhower.csv")
truman = pd.read_csv("Historical_Polls/truman.csv")
fdr = pd.read_csv("Historical_Polls/fdr.csv")

historical_polls = pd.merge(trump, obama['Approving'], left_index=True, right_index=True)
historical_polls = pd.merge(historical_polls, bush_w['Approving'], left_index=True, right_index=True)
historical_polls = pd.merge(historical_polls, clinton['Approving'], left_index=True, right_index=True)
historical_polls = pd.merge(historical_polls, bush_hw['Approving'], left_index=True, right_index=True)
historical_polls = pd.merge(historical_polls, reagan['Approving'], left_index=True, right_index=True)
historical_polls = pd.merge(historical_polls, carter['Approving'], left_index=True, right_index=True)
#historical_polls = pd.merge(historical_polls, ford['Approving'], left_index=True, right_index=True)
#historical_polls = pd.merge(historical_polls, nixon['Approving'], left_index=True, right_index=True)
#historical_polls = pd.merge(historical_polls, johnson['Approving'], left_index=True, right_index=True)
#historical_polls = pd.merge(historical_polls, kennedy['Approving'], left_index=True, right_index=True)
#historical_polls = pd.merge(historical_polls, eisenhower['Approving'], left_index=True, right_index=True)
#historical_polls = pd.merge(historical_polls, truman['Approving'], left_index=True, right_index=True)
#historical_polls = pd.merge(historical_polls, fdr['Approving'], left_index=True, right_index=True)
```

```python
historical_polls
historical_polls.columns=['Start Date', 'End Date', 'Trump', 'Disapproving', 'Unsure', 'Obama', 'Bush W', 'Clinton', 'Bush HW', 'Reagan', 'Carter']
historical_polls.reindex(index=historical_polls.index[::-1])
```

```python
plt.rcParams.update({'font.size': 22})
plt.plot(historical_polls['End Date'], historical_polls['Trump'], label = "Trump")
plt.plot(historical_polls['End Date'], historical_polls['Obama'], label = "Obama")
plt.plot(historical_polls['End Date'], historical_polls['Bush W'], label = "Bush W")
plt.plot(historical_polls['End Date'], historical_polls['Clinton'], label = "Clinton")
plt.plot(historical_polls['End Date'], historical_polls['Bush HW'], label = "Bush HW")
plt.plot(historical_polls['End Date'], historical_polls['Reagan'], label = "Reagan")
plt.plot(historical_polls['End Date'], historical_polls['Carter'], label = "Carter Approval")
plt.xlabel('First 4 Years of Presidency')
# Set the y axis label of the current axis.
plt.ylabel('Approval Rating')
plt.tick_params(
    axis='x',          
    which='both',      
    bottom=False,      
    top=False,         
    labelbottom=False)
# Set a title of the current axes.
plt.title('Historical Presidential Approval Ratings')
# show a legend on the plot
plt.legend()
plt.gcf().set_size_inches(20, 13)
plt.savefig("historical_polls.png")
# Display a figure.
plt.show()
```
