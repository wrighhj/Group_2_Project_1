```python
trump_data = "approval_ratings/trump_ratings.csv"
trump_ratings = pd.read_csv(trump_data)
trump_ratings.head(120)
```

```python
congressional_ballot = "approval_ratings/generic_polllist.csv"
general_ballot = pd.read_csv(congressional_ballot)
general_ballot.head(10)
```

```python
general_ballot['enddate'] = pd.to_datetime(general_ballot['enddate'])

general_ballot.dtypes

generic_ballot_2020 = general_ballot.loc[general_ballot['enddate'] > '2020-1-20']
generic_ballot_2020.head(100)

generic_ballot_2020.groupby('enddate').aggregate({'adjusted_dem': 'mean', 'adjusted_rep': 'mean'})

generic_ballot_2020.plot(x='enddate', y='adjusted_rep');

#generic_ballot_2020.reset_index(level=['enddate'], inplace=True)

#generic_ballot_2020.index.to_pydatetime()

#df = df[(df['date'] > '2000-6-1') & (df['date'] <= '2000-6-10')]
#general_ballot_clean.types()
#general_ballot_clean.plot();

#general_ballot_clean.loc[general_ballot_clean['enddate']]

```

```python
y_values = generic_ballot_2020['adjusted_rep']
x_values = generic_ballot_2020['enddate']
(slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)
regress_values = x_values * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
plt.scatter(x_values,y_values)
plt.plot(x_values,regress_values,"r-")
plt.annotate(line_eq,(6,10),fontsize=15,color="red")
plt.xlabel('Survey Enddate')
plt.ylabel('Approval')
plt.show()
```
